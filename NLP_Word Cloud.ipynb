```python
# A sample text snippet from an article on the web about UCU hosting the All Africa Horticulture Congress
# A remark made by Prof. Mushengyezi, the VC
Remark = '''This is a big win for Uganda, which has become one of the top destinations for international visitors including investors and tourists. 
Hosting AAHC 2028 also indicates growing confidence in UCU as a reputable university in Uganda and the region. 
We pledge to continue diligently serving humanity as a Centre of Excellence in the heart of Africa'''
```

```python
## Cleaning the Remark to remove all punctuations, numbers and special characters
import re

# removing all the special characters
# Just a template to show how it can be done Selectively
cleanRemark = re.sub(r'[?|$|.|!]',r' ',Remark)

# remove everything which is not Alphabets
cleanRemark = re.sub(r'[^a-z A-Z]',r' ',cleanRemark)  

# Convert the whole text to lowercase
cleanedRemark = cleanRemark.lower()     

# Delete any word which is less than 3-characters mostly those are stopwords
cleanRemark= re.sub(r'\b\w{1,3}\b', ' ', cleanRemark)

# Strip off extra spaces in the text
cleanRemark= re.sub(r' +', ' ', cleanRemark)
```

```python
# Plot the wordcloud
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from wordcloud import WordCloud, STOPWORDS

# Creating a custom list of stopwords
customStopwords=list(STOPWORDS) + ['Uganda','Africa','UCU','region']

wordcloudimage = WordCloud( max_words=50,
                            font_step=2 ,
                            max_font_size=500,
                            stopwords=customStopwords,
                            background_color='black',
                            width=1000,
                            height=720
                          ).generate(cleanRemark)

plt.figure(figsize=(15,7))
plt.imshow(wordcloudimage)
plt.axis("off")
plt.show()
```

```python
#What if we change the stop words to be anything short 
from wordcloud import WordCloud, STOPWORDS

# Creating a custom list of stopwords
customStopwords=list(STOPWORDS) + ['a','we','as','of']

wordcloudimage = WordCloud( max_words=50,
                            font_step=2 ,
                            max_font_size=500,
                            stopwords=customStopwords,
                            background_color='black',
                            width=1000,
                            height=720
                          ).generate(cleanRemark)

plt.figure(figsize=(15,7))
plt.imshow(wordcloudimage)
plt.axis("off")
plt.show()
```

```python
#What if we change the stop words to be anything short 
#And change the maximum words to 20
from wordcloud import WordCloud, STOPWORDS

# Creating a custom list of stopwords
customStopwords=list(STOPWORDS) + ['a','we','as','of']

wordcloudimage = WordCloud( max_words=20,
                            font_step=2 ,
                            max_font_size=500,
                            stopwords=customStopwords,
                            background_color='black',
                            width=1000,
                            height=720
                          ).generate(cleanRemark)

plt.figure(figsize=(15,7))
plt.imshow(wordcloudimage)
plt.axis("off")
plt.show()
```
