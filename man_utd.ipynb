```python
import sys
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import nltk
import requests
from bs4 import BeautifulSoup

# Ensure necessary NLTK data files are downloaded
nltk.download('punkt')
nltk.download('stopwords')

import requests
from bs4 import BeautifulSoup
import pandas as pd

markup = 'https://www.manutd.com/en/news/detail/casemiro-praise-for-busquets-dunga-kroos-modric-scholes-and-zidane'

from bs4 import BeautifulSoup
import requests
r=requests.get(markup)
soup=BeautifulSoup(r.content,"html.parser")
#
# Fetch text content from the URL
text = soup.find("div",{"class":"home-article-detail__content-panel"}).get_text() #<div class=kingster-single-article-content>

if text:
    # Preprocess the text
    text = text.lower()
    words = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.isalpha() and word not in stop_words]
    cleaned_text = ' '.join(words)

    # Generate the word cloud
    wordcloud = WordCloud(stopwords=stop_words, background_color='white').generate(cleaned_text)

    # Display the generated image
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
plt.show()
```
